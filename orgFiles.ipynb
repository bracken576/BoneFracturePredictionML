```python
import os
import shutil

# print("Elbow")
# folder = r"Bone-Fracture-Detection/Elbow"
# subfolders = [f.path for f in os.scandir(folder) if f.is_dir()]
# count = 0
# for sub in subfolders:
#     for f in (os.listdir(sub)):
#         if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
#             src = os.path.join(sub, f)
#             for image in (os.listdir(src)):
#                 if str(src)[-8:] == "negative":
#                     src2 = os.path.join(src, image)
#                     os.rename(src2, src+'\\image0'+str(count)+'.png')
#                     image = 'image0'+str(count)+'.png'
#                     src2 = src+'\\image0'+str(count)+'.png'
                    
#                     folder2 = r"Bone-Fracture-Detection/Elbow/Non-Fractured"
#                     shutil.move(src2, folder2)
                    
#                 else:
#                     src2 = os.path.join(src, image)
#                     os.rename(src2, src+'\\image0'+str(count)+'.png')
#                     image = 'image0'+str(count)+'.png'
#                     src2 = src+'\\image0'+str(count)+'.png'
                    
#                     folder2 = r"Bone-Fracture-Detection/Elbow/Fractured"
#                     shutil.move(src2, folder2)
                    
#                 count += 1
            
# for sub in subfolders:
#     if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
#         shutil.rmtree(sub)
      


# print("Hand")
# folder = r"Bone-Fracture-Detection/Hand"
# subfolders = [f.path for f in os.scandir(folder) if f.is_dir()]
# count = 0
# for sub in subfolders:
#     for f in (os.listdir(sub)):
#         if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
#             src = os.path.join(sub, f)
#             for image in (os.listdir(src)):
#                 if str(src)[-8:] == "negative":
#                     src2 = os.path.join(src, image)
#                     os.rename(src2, src+'\\image0'+str(count)+'.png')
#                     image = 'image0'+str(count)+'.png'
#                     src2 = src+'\\image0'+str(count)+'.png'
                    
#                     folder2 = r"Bone-Fracture-Detection/Hand/Non-Fractured"
#                     shutil.move(src2, folder2)
                    
#                 else:
#                     src2 = os.path.join(src, image)
#                     os.rename(src2, src+'\\image0'+str(count)+'.png')
#                     image = 'image0'+str(count)+'.png'
#                     src2 = src+'\\image0'+str(count)+'.png'
                    
#                     folder2 = r"Bone-Fracture-Detection/Hand/Fractured"
#                     shutil.move(src2, folder2)
                    
#                 count += 1
            
# for sub in subfolders:
#     if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
#         shutil.rmtree(sub)
       
       

print("Shoulder")
folder = r"Bone-Fracture-Detection/Shoulder"
subfolders = [f.path for f in os.scandir(folder) if f.is_dir()]
count = 0
for sub in subfolders:
    for f in (os.listdir(sub)):
        if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
            src = os.path.join(sub, f)
            for image in (os.listdir(src)):
                if str(src)[-8:] == "negative":
                    src2 = os.path.join(src, image)
                    os.rename(src2, src+'\\image0'+str(count)+'.png')
                    image = 'image0'+str(count)+'.png'
                    src2 = src+'\\image0'+str(count)+'.png'
                    
                    folder2 = r"Bone-Fracture-Detection/Shoulder/Non-Fractured"
                    shutil.move(src2, folder2)
                    
                else:
                    src2 = os.path.join(src, image)
                    os.rename(src2, src+'\\image0'+str(count)+'.png')
                    image = 'image0'+str(count)+'.png'
                    src2 = src+'\\image0'+str(count)+'.png'
                    
                    folder2 = r"Bone-Fracture-Detection/Shoulder/Fractured"
                    shutil.move(src2, folder2)
                    
                count += 1
            
for sub in subfolders:
    if not(sub[-13:] == "Non-Fractured" or sub[-9:] == "Fractured"):
        shutil.rmtree(sub)
```
